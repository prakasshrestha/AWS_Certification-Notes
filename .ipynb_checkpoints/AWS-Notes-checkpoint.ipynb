## 1. 🔐 AWS Shared Responsibility Model (EC2 example)

### 📌 Core idea

**Security & Compliance = AWS + Customer (shared)**

तर जिम्मेवारी कुन तहसम्म कसको हो भन्ने कुरा service अनुसार फरक पर्छ।

---

### ☁️ AWS = Security OF the Cloud

AWS ले cloud को भित्री infrastructure सुरक्षित गर्छ।

**AWS को जिम्मेवारी:**
- Global infrastructure (regions, AZs)
- Data centers (physical security)
- Hardware, networking
- Host OS
- Virtualization layer (hypervisor)
- AWS managed services को underlying patching

**👉 Exam मा देखिने शब्द:**
- Configuration management for AWS global infrastructure ❌ (AWS)
- Availability Zone (AZ) infrastructure management ❌ (AWS)
- Patching/fixing flaws within AWS infrastructure ❌ (AWS)

*यी सबै AWS को responsibility हुन्।*

---

### 🧑‍💻 Customer = Security IN the Cloud

Customer ले EC2 भित्र के चल्छ त्यो manage गर्छ।

**Customer को जिम्मेवारी (EC2 मा):**
- ✅ Guest Operating System patching (Linux/Windows update)
- ✅ Security updates (yum, apt, Windows Update)
- ✅ Installed applications & utilities
- ✅ EC2 Security Group configuration
- ✅ IAM usage inside workload
- ✅ Data encryption (depending on use case)

> **👉 Key exam line (important)**
>
> Customers that deploy an Amazon EC2 instance are responsible for the management of the guest operating system (including updates and security patches).

---

### ❓ Exam Question pattern (100% aauxa)

**Q: Who is responsible for patching the guest OS on an EC2 instance?**
✅ Customer

**Q: Who patches the host OS / hypervisor?**
✅ AWS

**Q: Who manages Availability Zone infrastructure?**
✅ AWS

---

### 🧠 Easy memory trick (Exam hack)

```
EC2 = OS dekhi mathi → Customer
EC2 = OS muni → AWS

Application        → Customer
Guest OS           → Customer  ✅ (THIS QUESTION)
Security Group     → Customer
-------------------------------
Hypervisor         → AWS
Host OS            → AWS
Physical hardware  → AWS
AZ / Region        → AWS
```

## 🔖 Amazon DynamoDB (Storage)

> **Amazon DynamoDB = NoSQL | Serverless | Flexible Schema**

### 🧠 Core Idea

- Key-Value + Document database
- Schema-less (Flexible schema)
- Auto-scaling, high performance, fully managed

### 📦 Data Model

- **Table**
- **Item** = row
- **Attribute** = column
- **Primary Key:**
  - Partition Key
  - Partition + Sort Key

### 🚀 Features

- Serverless (instance / server manage गर्नु पर्दैन)
- Massive scale (millions TPS)
- Pay-as-you-go
- Global Tables (multi-region)

### ❌ Not suitable for

- Complex JOIN
- Heavy SQL queries
- Strict relational schema

### ✅ Best Use Cases

- User profile
- Session store
- IoT data
- Gaming leaderboard
- Shopping cart

> **📝 Exam Magic Line**
>
> "Flexible schema + high scale = DynamoDB"

## 🌱 AWS Elastic Beanstalk (Compute service)

> **AWS Elastic Beanstalk = Platform as a Service (PaaS) | Fully Managed | Auto-scaling**

### 🧠 Core Idea

Application deployment र infrastructure management को लागि fully managed service

### EC2 vs Elastic Beanstalk Comparison

| ❌ EC2 प्रयोग गर्दा | ✅ Elastic Beanstalk प्रयोग गर्दा |
|---|---|
| server setup | code upload गर्नुहोस् |
| software install | AWS ले सबै manage गर्छ |
| scaling setup | automatic scaling |
| load balancer configure | automatic load balancer |
| 👉 सबै आफैं गर्नुपर्छ | 👉 server, scaling, load balancer automatically |

### ✅ Elastic Beanstalk कहिले प्रयोग गर्ने?

- Web app छ (Java, Node.js, Python, PHP आदि)
- Server manage गर्न चाहनुहुन्न
- Fast deployment चाहनुहुन्छ

> **🏠 Real-life analogy**
>
> **EC2** = आफैं घर बनाएर बस्ने
> **ECS/EKS** = apartment system manage गर्ने
> **Elastic Beanstalk** = fully managed hotel मा बस्ने
>
> 👉 तपाईं केवल बस्नुहुन्छ
> 👉 service अरूले manage गर्छ

### 🚀 Key Features

- **Automatic scaling** – traffic अनुसार auto-scale हुन्छ
- **Load balancing** – automatic setup हुन्छ
- **Health monitoring** – application health track गर्छ
- **Multiple languages** – Java, .NET, PHP, Node.js, Python, Ruby, Go, Docker
- **Version management** – multiple versions deploy र rollback गर्न सकिन्छ

### 📦 What Elastic Beanstalk Manages

- EC2 instances
- Load balancer (ELB)
- Auto-scaling groups
- Security groups
- Database (RDS) – optional
- Monitoring (CloudWatch)

### ❌ What You Still Control

- Application code
- Configuration settings
- Environment variables

> **📝 Exam Magic Line**
>
> "Deploy code without managing infrastructure = Elastic Beanstalk"

> **🔚 Final Exam Tip 💡**
>
> **यदि question मा यी शब्द देख्यौ भने 👇**
>
> *PaaS, deploy code, no infrastructure management, automatic scaling*
>
> **👉 Answer = Elastic Beanstalk 🎯**

## 🔐 AWS CloudHSM

> **AWS CloudHSM = Cloud मा उपलब्ध dedicated Hardware Security Module (HSM)**
>
> 👉 Encryption key generate, store र use गर्नका लागि।

### 🔑 Core Concept

| Feature | CloudHSM |
|---------|----------|
| **HSM Type** | Dedicated (Single Customer) |
| **Security Level** | FIPS 140-2 Level 3 |
| **Key Control** | Full Customer Control |
| **Managed By** | AWS (hardware, patch, HA) |
| **Used For** | High-security encryption |

### 🎯 When to Use CloudHSM?

- ✔ Regulatory compliance (Bank, Gov, Healthcare)
- ✔ Own crypto library use गर्नुपर्छ
- ✔ Key ownership fully control गर्नुपर्छ
- ✔ Dedicated hardware चाहिन्छ

### ❌ When NOT to Use?

| Need | Use This Instead |
|------|------------------|
| Simple encryption | **AWS KMS** |
| Password storage | **Secrets Manager** |
| Cost/security suggestion | **Trusted Advisor** |

### 🔥 CloudHSM vs KMS Quick Trick

| Question Clue | Answer |
|---------------|--------|
| Dedicated HSM | **CloudHSM** |
| FIPS Level 3 | **CloudHSM** |
| Easy key management | **KMS** |
| Rotate DB password | **Secrets Manager** |

---

## 🧠 Amazon Neptune (Storage)

> **Amazon Neptune = Graph Database Service**
>
> यो विशेष गरी सम्बन्ध (relationship) धेरै भएको data का लागि बनाइएको हो।

### 📌 Graph Database भनेको के हो?

**Graph DB मा data यस्तो हुन्छ:**
- **Node** (वस्तु) → मान्छे, product, स्थान
- **Edge** (सम्बन्ध) → friend, bought, connected
- **Property** → उमेर, मूल्य, मिति

👉 Table जस्तो होइन
👉 Connection खोज्न धेरै छिटो

### 🎯 Neptune कहिले प्रयोग गर्ने?

- ✔ Social network
- ✔ Recommendation engine
- ✔ Fraud detection
- ✔ Knowledge graph
- ✔ Network analysis

### 🔥 मुख्य विशेषता

- Fully managed (AWS ले manage गर्छ)
- Highly available
- Automatic backup
- Relationship query मा धेरै fast
- Gremlin र SPARQL support गर्छ

### ❌ कहिले प्रयोग नगर्ने?

| जरुरत | प्रयोग गर्ने service |
|-------|-------------------|
| Simple key-value | **DynamoDB** |
| Table based SQL | **RDS** |
| Big analytics | **Redshift** |

### 📊 Graph DB vs NoSQL (DynamoDB) तुलना

| कुरा | Graph DB (Neptune) | NoSQL (DynamoDB) |
|-----|-------------------|------------------|
| **Data structure** | Node & Edge | Key-Value / Document |
| **Relationship query** | धेरै fast | गाह्रो |
| **Massive scale** | राम्रो | अझै राम्रो |
| **Simple lookup** | ठीक | सबैभन्दा राम्रो |
| **Recommendation logic** | Excellent | Complicated |

## 🔐 AWS Identity and Access Management (IAM)

> **IAM भनेको AWS भित्र कोले के गर्न पाउँछ? भनेर control गर्ने service हो।**
>
> 👉 Permission control system हो।

### 1️⃣ IAM Entities – Full Explanation

#### 👤 1. IAM User

**📌 के हो?**
व्यक्ति वा application को लागि individual identity।

**🧠 Example**
- Company को DevOps engineer
- Developer
- Application script

**🔑 User सँग हुन सक्छ:**
- Username & Password (Console login)
- Access Key ID & Secret Access Key (CLI/API login)

---

#### 👥 2. IAM Group

**📌 के हो?**
धेरै user लाई एउटै permission दिन group बनाइन्छ।

**🧠 Example**
- Developers group → सबै developer लाई एउटै permission

**👉 Key Points:**
- Group मा Access Key हुँदैन
- Group मा Policy attach हुन्छ

---

#### 🎭 3. IAM Role

**📌 के हो?**
Temporary access दिन प्रयोग हुने identity।

**🧠 Example**
- EC2 ले S3 access गर्न
- Cross-account access
- Lambda function permission

**👉 Key Points:**
- Role मा Access Key permanently हुँदैन
- Temporary credentials generate हुन्छ

---

#### 📜 4. IAM Policy

**📌 के हो?**
Permission define गर्ने JSON document।

**🧠 Example**
- S3 read only
- EC2 full access
- Deny delete action

**👉 Policy = "के गर्न पाइन्छ?"**

---

### 📊 Quick Table

| Entity | Login गर्न सक्छ? | Access Key हुन्छ? |
|--------|-----------------|-----------------|
| User | Yes | Yes |
| Group | No | No |
| Role | Direct login होइन | Temporary |
| Policy | No | No |

---

### 2️⃣ Login Types Comparison

#### 🌐 1. Console Login
👉 Browser बाट login
- Username
- Password
- MFA (optional)

#### 💻 2. Programmatic Login
👉 CLI, SDK, Script बाट login
- Access Key ID
- Secret Access Key

> **📌 Important**
>
> **Access Key ID + Secret Access Key**
> 👉 Permanent रूपमा IAM User सँग attach हुन्छ
>
> **Role ले temporary key generate गर्छ।**

---

### 3️⃣ Exam Trick Explanation 🔥

**Exam मा question यस्तो आउँछ:**
*"Access Key ID and Secret Access Key are tied to which IAM entity?"*

**Trick:**
- Group → No login
- Policy → Permission only
- Role → Temporary credential
- **User → Permanent access key** ✅

### 🎯 5-Second Memory Trick

| यदि question मा | सोच्ने |
|----------------|--------|
| Username/Password | **IAM User** |
| Access Key | **IAM User** |
| Temporary credential | **IAM Role** |
| Permission document | **Policy** |

---

### 🇳🇵 Nepali Use Cases

**Scenario 1: Developer Access**
- **Need:** Developer लाई AWS Console र CLI दुवै access चाहियो
- **Solution:** IAM User बनाउने with Username/Password (Console) + Access Keys (CLI)

**Scenario 2: Application on EC2**
- **Need:** EC2 running application लाई DynamoDB access चाहियो
- **Solution:** IAM Role create गरेर EC2 मा attach गर्ने

**Scenario 3: Team Management**
- **Need:** 10 जना developer लाई same permission चाहियो
- **Solution:** IAM Group बनाउने र सबै user लाई group मा add गर्ने

> | Temporary Credential | **IAM Role** |
> | Console Login (Browser) | **IAM User** |
> | Permission Document | **IAM Policy** |
> | Group Users Together | **IAM Group** |
> | EC2 needs S3 access | **IAM Role** |
>
> **👉 IAM = Who can do What 🎯**

## 📩 Amazon MQ

> **Amazon MQ = Managed Message Broker Service**
>
> यसले **Apache ActiveMQ** र **RabbitMQ** लाई AWS मा manage गरेर चलाउन सजिलो बनाउँछ।

### 🧠 Message Broker के हो?

Message Broker भनेको system हरू बीच message पठाउने बिचको माध्यम हो।

```
App A → Message Broker → App B
```

### 🎯 किन चाहिन्छ?

यदि दुई system:
- फरक programming language मा छन्
- Direct connection risky छ
- Reliable message delivery चाहिन्छ

**👉 तब Message Broker प्रयोग हुन्छ।**

### 📦 Amazon MQ के गर्छ?

- Server provisioning AWS ले गर्छ
- Setup AWS ले गर्छ
- Patch/maintenance AWS ले गर्छ
- Backup/HA manage गर्छ

### 🏢 कहिले प्रयोग गर्ने?

- पुरानो application छ
- Already ActiveMQ वा RabbitMQ प्रयोग भइरहेको छ
- Code rewrite गर्न चाहँदैनौ
- Cloud मा migrate गर्न चाहन्छौ

### ❓ Amazon MQ vs SQS vs SNS

| Service | Best For |
|---------|----------|
| **Amazon MQ** | Existing ActiveMQ/RabbitMQ migrate |
| **SQS** | Simple queue (new cloud-native app) |
| **SNS** | Notification / Pub-Sub |

### 🎯 Exam Trick

**If question says:**
- Existing messaging system / Industry standard protocol / No code rewrite / ActiveMQ / RabbitMQ

**👉 Answer = Amazon MQ**

**If question says:**
- New cloud-native app / Serverless messaging

**👉 Answer = SQS or SNS**

#### 🔹 1️⃣ AWS Step Functions के हो?

👉 AWS को Workflow Orchestration Service हो।
👉 धेरै AWS services (Lambda, Batch, ECS, SNS, DynamoDB आदि) लाई
step-by-step sequence मा connect गरेर run गराउँछ।
### 🔹 Real Architecture Example

**📦 E-commerce Order Processing with Step Functions:**

```
API Gateway (Order placed)
    ↓
Step Functions START
    ↓
State 1: Lambda (Validate Order)
    ↓
State 2: Choice (Valid?)
    ↓
    ├─→ YES → State 3: Lambda (Process Payment)
    │              ↓
    │         State 4: Choice (Payment Success?)
    │              ↓
    │         ├─→ YES → State 5: Lambda (Update Inventory)
    │         │            ↓
    │         │       State 6: Lambda (Send Confirmation)
    │         │            ↓
    │         │       SUCCESS
    │         │
    │         └─→ NO → State 7: Lambda (Refund)
    │                      ↓
    │                  State 8: SNS (Notify Customer)
    │                      ↓
    │                  FAIL
    │
    └─→ NO → State 9: Lambda (Invalid Order Notification)
                 ↓
             FAIL
```

**👉 सबै steps Step Functions ले manage गर्छ**  
**👉 Retry, error handling automatic**  
**👉 Visual console मा देख्न सकिन्छ**

### 🔹 📖 Complete Exam Notes (Enhanced)

#### 🔹 Flow Concept (Basic Diagram)

```
Start
  ↓
Task (Lambda)
  ↓
Choice (Condition check)
   ├── Yes → Task 2
   │          ↓
   │      Success
   │
   └── No → Fail
```

**👉 यो सबै JSON मा define हुन्छ (State Machine definition)**  
**👉 AWS Console मा visual editor पनि छ**

### 🔹 Standard vs Express (Exam Important ⚡)

| Feature | Standard | Express |
|---------|----------|---------|
| **Duration** | Long-running (up to 1 year) | Short/high volume (5 min) |
| **Execution rate** | 2,000 per second | 100,000 per second |
| **Cost** | Per state transition | Per execution + duration |
| **Use case** | Business workflow | High-speed processing |
| **Visual history** | Complete execution history | CloudWatch Logs only |
| **Exactly-once** | Yes | At-least-once |

**👉 Exam मा धेरैजसो scenario → Standard Workflows**  
**👉 IoT, streaming, high-throughput → Express Workflows**

### 🔹 Built-in Features (Why Step Functions is Powerful)

**✔ Automatic retry**
- State level मा retry count configure गर्न सकिन्छ
- Exponential backoff support

**✔ Error handling**
- Catch error state मा जान सकिन्छ
- Fallback logic define गर्न सकिन्छ

**✔ Visual workflow**
- AWS Console मा graph देख्न सकिन्छ
- Real-time execution tracking

**✔ JSON-based definition**
- Infrastructure as Code

**✔ Parallel execution**
- धेरै task एकै साथ run गर्न सकिन्छ
- Faster processing

### 🔹 Additional State Types (Complete List)

| State Type | Function | Example Use |
|------------|----------|-------------|
| **Task** | Service call गर्ने | Lambda invoke, Batch job submit |
| **Choice** | Conditional branching | If-else logic |
| **Wait** | Delay execution | 30 seconds wait, wait until time |
| **Parallel** | Multiple branches simultaneously | Process video + generate thumbnail |
| **Map** | Array items iterate | Process 100 images one-by-one |
| **Pass** | Data transformation | Add/modify state data |
| **Succeed** | Successful termination | End with success |
| **Fail** | Failure termination | End with error |

### 🔹 Retry & Error Handling Example

# AWS Global Accelerator

## 1) परिचय

    AWS Global Accelerator एक networking service हो जसले प्रयोगकर्ताको ट्राफिकलाई AWS को global network मार्फत सबैभन्दा नजिकको edge location हुँदै छिटो र स्थिर रूपमा application सम्म पुर्‍याउँछ। यसले performance सुधार गर्छ र latency घटाउँछ।

## 2) मुख्य विशेषता (Key Features)
    🔹 Static IP Address
    
    2 वटा स्थिर (static) IP address प्रदान गर्छ।
    
    DNS change नगरी backend परिवर्तन गर्न सकिन्छ।
    
    🔹 Performance Improvement
    
    AWS को private global network प्रयोग गर्छ।
    
    सार्वजनिक इन्टरनेटभन्दा छिटो र स्थिर route प्रयोग हुन्छ।
    
    🔹 Low Latency
    
    प्रयोगकर्ताको नजिकको edge location मा ट्राफिक लिन्छ।
    
    प्रतिक्रिया समय घट्छ।
    
    🔹 Automatic Failover
    
    यदि एउटा endpoint down भयो भने अर्को healthy endpoint मा ट्राफिक पठाउँछ।
    
    Health check स्वतः हुन्छ।
    
    🔹 Traffic Distribution
    
    विभिन्न Region मा रहेका endpoints बीच ट्राफिक सन्तुलन (load balancing) गर्छ।

## 3) कसरी काम गर्छ?

    प्रयोगकर्ताले application को static IP मा request पठाउँछ।
    
    Request नजिकको AWS edge location मा पुग्छ।
    
    त्यहाँबाट AWS global backbone network प्रयोग गरी target Region मा पठाइन्छ।
    
    Healthy endpoint (जस्तै Load Balancer वा EC2) मा request पुग्छ।

## 4) Supported Endpoints

    Application Load Balancer
    
    Network Load Balancer
    
    EC2 Instance
    
    Elastic IP address

## 5) फाइदाहरू (Advantages)

    विश्वव्यापी performance सुधार
    
    High availability
    
    Automatic failover
    
    स्थिर IP address
    
    Multi-region support

## 6) Global Accelerator vs Route 53 vs CloudFront

    Amazon Route 53 → DNS based routing service
    
    Amazon CloudFront → Content Delivery Network (CDN)
    
    AWS Global Accelerator → Network layer (Layer 4) traffic acceleration service

## 7) परीक्षामुखी छोटो नोट

    AWS Global Accelerator ले static IP address प्रदान गरी AWS को global network प्रयोग गरेर user traffic लाई नजिकको edge location हुँदै healthy endpoint मा पठाउँछ, जसले latency घटाउँछ र availability बढाउँछ।

>>>>  

    User
      ↓
    Static IP (Global Accelerator)
      ↓
    Nearest AWS Edge Location
      ↓
    AWS Global Network (Private Backbone)
      ↓
    Region
      ↓
    Endpoint (ALB / NLB / EC2)

* ※　AWS Global Accelerator HTTP मात्र होइन, TCP/UDP आधारित non-HTTP application हरूका लागि पनि उपयुक्त छ, किनकि यसले Layer 4 मा काम गर्छ र performance तथा availability सुधार गर्छ।

# ✅ Amazon Snowball भनेको के हो?

**Amazon Snowball** भनेको  
👉 **Extremely large amount को data (TB / PB level)**  
👉 **Internet प्रयोग नगरी**  
👉 **Physical device मार्फत AWS cloud मा transfer गर्ने service** हो।

---

## 🔑 सरल शब्दमा:
- Data धेरै ठूलो छ
- Internet slow / expensive छ
- Snowball device AWS ले पठाउँछ
- Data copy गर्छौं
- Device AWS लाई फिर्ता पठाउँछौं
- AWS ले data S3 / Glacier मा upload गर्छ

🧠 **Snowball = Data shipment service**

---

# 🤔 Real Life Example (दिमागमा सेट गराउने)

मानौँ:
- Company सँग **200 TB data** छ
- Internet speed slow छ
- Upload गर्न महिना / वर्ष लाग्छ

### 🟢 Internet बाट upload गर्दा:
- Time धेरै लाग्छ
- Cost धेरै
- Network failure risk

### 🟢 Amazon Snowball प्रयोग गर्दा:
- AWS बाट **Snowball device courier** हुन्छ
- Company ले data **local server बाट Snowball मा copy** गर्छ
- Device **AWS लाई return** गर्छ
- AWS ले data **cloud मा upload** गर्छ

# 🧊 Amazon Snowball Family

## 1️⃣ Snowball Edge Storage Optimized
- Large storage capacity
- TB/PB data transferका लागि
- Limited compute capability

## 2️⃣ Snowball Edge Compute Optimized
- Data transfer + **powerful compute**
- Edge processing (IoT, ML, etc.)

## 3️⃣ Snowmobile (For exam awareness)
- **Exabyte (EB) level data**
- Truck भित्र data center जस्तै device

🧠 *Snowball = TB*  
🧠 *Snowmobile = EB*

# 🧱 Amazon Snowball का Main Components

## 1️⃣ Snowball Device
- Physical hardware
- Storage + network interface
- Encryption built‑in

## 2️⃣ Snowball Client
- Software tool
- Data copy गर्न प्रयोग हुन्छ
- Command line based

## 3️⃣ AWS S3 / Glacier
- Final storage destination
